```python
""" Implementing softmax in TF """
def softmax(x):
    sf = tf.nn.softmax(tf.cast(x,tf.float32))
    print("-------",sf[0],sum(sf[0]))

def softmax2(x):
    result = []
    for row in x:
        ezi = np.exp(row)
        ezj = np.sum(ezi)
        sf = ezi/ezj
        result.append(sf)
        print("Ans by Softmax2 = ",sf,np.sum(sf))

def pool_ex11():
    random_array = np.random.randint(30,size=(4,7))
    x1 = tf.constant(random_array)
    softmax(x1)
    x2 = tf.constant([[-1,-1,5]])
    softmax(x2)
    x3 = tf.constant([[1,1,5]])
    softmax(x3)


    x4 = tf.constant([[8,5,0]])
    softmax2(x4)        # expected ans = [0.9523 , 0.0474 , 0.0003] if you will sum you will get 1



pool_ex11()
```
